```python
import pandas as pd
```

```python
raw_data = [["Zenglongke", 5],
            ["Ouyangkang", 14],
            ["Wuyaoxuan", 12],
            ["Heluqi", 10],
            ["Limeiyu", 4],
            ["Xieyuwei", 3],
            ["Zhouyupeng", 3],
            ["Xiamuqing", 1],
            ["Zhangyifan", 2],
            ["Chenxiaoya", 2],
            ["Lizhuoran", 1]]
data = pd.DataFrame(raw_data, columns=['Name', 'Times'])
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(16, 16), subplot_kw=dict(aspect='equal'))
wages, text = plt.pie((data.Times / data.Times.sum()).tolist(), labels=data.Name)
ax.legend(wages, data.Name, loc='upper right')
```
