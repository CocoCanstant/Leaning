```python
import pandas as pd

data = pd.read_csv('/mnt/c/Users/ouyangkang/Desktop/multi_infect_to_human.csv')
```

```python
data
```

```python
trans_family = pd.DataFrame(columns=['Type','Ratio'])
n = 0
for family in list(data['VirusFamily']):
    multi = round(int(data[data['VirusFamily']== family]['IMFE_Infect_human'])/int(data[data['VirusFamily']== family]['Infect_MultiFamily_Except_hominidae']),3)
    if int(data[data['VirusFamily']== family]['OF_Infect_human']) != 0:
        one = round(int(data[data['VirusFamily']== family]['OF_Infect_human'])/int(data[data['VirusFamily']== family]['One_Family']),3)
    else:
        one = 0
    trans_family.loc[(n*2)+0] = ['Trans_Mulit',multi]
    trans_family.loc[(n*2)+1] = ['One',one]
    n += 1
```
