```python
import pandas as pd
```

```python
data = pd.read_csv("/mnt/d/dailywork/Mouse_brain/diffpeak_annotation.csv")
```

```python
data = data[data.Log2FC > 0]
data.head(5)
```

```python
data.drop_duplicates(subset=["X", "CT_I"], inplace=True)
```

```python
last_data = data.pivot(index="X", columns="CT_I", values="Log2FC")
```

```python
last_data.head(5)
print(last_data.shape)
```

```python
last_data.index.name = None
last_data.columns.name = None
last_data
```

```python
last_data[last_data.notna()] = 1
last_data.fillna(0, inplace=True)
```

```python
last_data.to_csv("/mnt/d/dailywork/Mouse_brain/upset_0430.csv")
```
