```python
from pathlib import Path
import pandas as pd
```

```python
data_dir = "/mnt/d/dailywork/Mouse_brain/fig_2B_raw_file"
```

```python
integrate_data = pd.DataFrame()

for path in Path(data_dir).iterdir():
    
    data = pd.read_csv(path, index_col=0)
    cell_type = path.name.split('_')[0]
    data[cell_type] = 1
    data = data[[cell_type]].T

    integrate_data = pd.concat([integrate_data, data])
```

```python
integrate_data = integrate_data.T
integrate_data.head()
```

```python
integrate_data.fillna(0, inplace=True)
integrate_data.head()
```

```python
len(integrate_data.index.tolist()) == len(set(integrate_data.index.tolist()))
```

```python
integrate_data.shape
```

```python
integrate_data.to_csv('/mnt/d/dailywork/Mouse_brain/integrated_data.csv')
```
