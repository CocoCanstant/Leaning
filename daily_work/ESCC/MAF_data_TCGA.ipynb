```python
import pandas as pd
```

```python
data = pd.read_csv("/mnt/d/dailywork/01_Smear_seq/cancer/eso_squamou_cancer/TCGA_eso_data/cohortMAF.2024-10-18", sep = "\t")
```

```python
print(data.shape)
data.head()
```

```python
data = data.dropna(axis=1, how="all")
data.shape
```

```python
len(set(data.case_id))
```

```python
sur_data = pd.read_csv("/mnt/d/dailywork/01_Smear_seq/cancer/eso_squamou_cancer/TCGA_eso_data/overall-survival-plot.2024-09-26.tsv", sep="\t")
```

```python
sur_data = sur_data[["id", "time", "censored"]]

sur_data.columns = ["id", "time", "overall_survival"]
sur_data['overall_survival'] = [1 if _ else 0 for _ in sur_data['overall_survival']]
sur_data.head()
```

```python
data.rename(columns={"case_id": "id"}, inplace=True)
```

```python
maf_data = pd.merge(data, sur_data, on="id")
```

```python
maf_data.head()
```

```python
maf_data.dropna(subset=["time"]).shape == maf_data.dropna(subset=["overall_survival"]).shape
```

```python
maf_data.dropna(subset=["time"], inplace=True)
```

```python
maf_data.to_csv("/mnt/d/dailywork/01_Smear_seq/cancer/eso_squamou_cancer/TCGA_eso_data/maf_with_survival_data.csv", index=None)
```
