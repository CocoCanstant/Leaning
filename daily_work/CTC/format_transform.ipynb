```python
# cat Y00724E4.txt | awk -F ' ' '{print $NF}' > /mnt/c/Users/ouyangkang/Desktop/tmp_Y007.txt
import pandas as pd
data = pd.read_csv('/mnt/c/Users/ouyangkang/Desktop/tmp_Y007.txt', header=None)
data.head()

from pathlib import Path
list_1 = []
list_2 = []

for i in data[0]:
    i = Path(i)
    file = i.parts[-1]
    judge_element = file.split('_')[-1].strip('.fq.gz')
    if judge_element == '1':
        list_1 += [i]  
    else:
        list_2 += [i]

print(len(list_1))
print(len(list_2))

final_df = pd.DataFrame({'FQ1': list_1,
             'FQ2': list_2})
final_df.to_csv('/mnt/c/Users/ouyangkang/Desktop/tmp_df.csv', index=None)
```
