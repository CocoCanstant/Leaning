```python
import pandas as pd
```

```python
day28 = pd.read_csv("/mnt/d/dailywork/Mouse_brain/GWAS/balanced_gwas/all_h2_results_day28.txt", sep="\t")
naive = pd.read_csv("/mnt/d/dailywork/Mouse_brain/GWAS/balanced_gwas/all_h2_results.txt", sep="\t")
```

```python
day28["celltype"] = day28.Category.str.split("_").str[0]
naive["celltype"] = naive.Category.str.split("_").str[0]
z_day28 = day28.pivot(index="pheno", columns="celltype", values="Coefficient_z-score")
z_naive = naive.pivot(index="pheno", columns="celltype", values="Coefficient_z-score")
```

```python
z_day28.index.name = None
z_naive.index.name = None
```

```python
z_day28.columns = [i + "Day28" for i in z_day28.columns]
```

```python
z_naive.columns = [i + "Naive" for i in z_naive.columns]
```

```python
z_day28
```

```python
final_df = pd.concat([z_day28, z_naive], axis=1)
```

```python
celltype = list(set(final_df.columns.str.split("-").str[0]))
timepoint = ["Naive", "Day28"]
for ct in celltype:
    tmp_df = final_df[[ct + "-" + timepoint[0], ct + "-" + timepoint[1]]]
    tmp_df.to_csv("/mnt/d/dailywork/Mouse_brain/GWAS/balanced_gwas/" + ct + "result.csv")
```
