```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import gseapy as gp
```

```python
mgl = pd.read_csv('/mnt/c/Users/ouyangkang/Desktop/go_activated_mgl_selected.csv')
mac = pd.read_csv('/mnt/c/Users/ouyangkang/Desktop/go_acticate_mac_slected.csv')
mgl['group'] = "MGL"
mac['group'] = "MAC"
```

```python
plot_df = pd.concat([mac, mgl], ignore_index=True)
```

```python
plot_df
```

```python
plot_df = plot_df.rename(columns={'Description': 'Term', 'p.adjust': 'Adjusted P-value'})
plot_df.head()
```

```python
gp.barplot(
    plot_df,
    figsize=(6, 8),
    title="Different function of DEGs between MGL and MAC",
    group="group",
    color=["b", "r"],
    column="Adjusted P-value",
    ofname = "activated_MGL-MAC_diff_function.pdf"
)
```
