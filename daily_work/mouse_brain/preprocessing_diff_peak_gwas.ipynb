```python
import pandas as pd
```

```python
data = pd.read_csv("/mnt/c/Users/ouyangkang/Documents/WeChat Files/wxid_btots2523c6z22/FileStorage/File/2024-06/diffpeaks.csv")
data = data[["X", "Time_CT_I", "Log2FC"]]
data = data[data.Time_CT_I.str.contains("Day 28")]
data = data[data.Log2FC > 0]
print(data.shape)
```

```python
ct_list = ['DGGRC', 'NGNBL', 'FB', 'MSN', 'MEGLU', 'MGL', 'OPC', 'PEP', 'AC', 'MAC', 'TEINH', 'CHOR', 'EPEN', 'TEGLU', 'OLG', 'EN']
for ct in ct_list:
    tmp_data = data[data.Time_CT_I.str.contains(ct)]
    # check whether data contain duplications
    if tmp_data.shape == tmp_data.drop_duplicates(["X"]).shape :

        tmp_data["Chromosome"] = tmp_data.X.str.split("_").str[0]
        tmp_data["Start"] = tmp_data.X.str.split("_").str[1]
        tmp_data["End"] = tmp_data.X.str.split("_").str[2]
        tmp_data = tmp_data[~tmp_data["End"].str.contains(".1")]
        tmp_data = tmp_data[["Chromosome", "Start", "End"]]
        tmp_data.to_csv("/mnt/c/Users/ouyangkang/Desktop/tmp_mouse/diff_peak/" + ct + "peak.bed", sep="\t", index=None)
    else:
        # drop duplications
        tmp_data.drop_duplicates(["X"], inplace=True)

        tmp_data["Chromosome"] = tmp_data.X.str.split("_").str[0]
        tmp_data["Start"] = tmp_data.X.str.split("_").str[1]
        tmp_data["End"] = tmp_data.X.str.split("_").str[2]
        tmp_data = tmp_data[~tmp_data["End"].str.contains(".1")]
        tmp_data = tmp_data[["Chromosome", "Start", "End"]]
        tmp_data.to_csv("/mnt/c/Users/ouyangkang/Desktop/tmp_mouse/diff_peak/" + ct + "peak.bed", sep="\t", index=None)
```
