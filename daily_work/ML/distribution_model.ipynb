```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# average list
mu_list = [0, 0, -1, 1]
# standard deviation list
sigma_list = [1, 2, 2, 2]
# color list
color_list = ['red', 'orange', 'lightseagreen', 'blue']
x = np.linspace(-5, 5, 100)

for i in range(4):
    # calculate the value of probability density function of normal distribution
    y = norm.pdf(x, mu_list[i], sigma_list[i])

    # plot normal distribution curve
    plt.plot(x, y, label=f'μ={mu_list[i]}, σ={sigma_list[i]}', color=color_list[i])

plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normal Distribution Curve')
plt.legend()
# plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# data generate
lambda_param = 5 # λ is the paramater of poisson distribution
x = np.arange(0, 100)
y = poisson.pmf(x, lambda_param) # calculate the value of probability mass function of poisson distribution

# plot bar figure
plt.bar(x, y, align='center', alpha=0.7, label='Poisson Distribution')

# plot curve figure
plt.plot(x, y, marker='o', color='red', linestyle='--', linewidth=2, markersize=5, label='Curve')

plt.xlabel('x')
plt.ylabel('Probability')
plt.title('Poisson Distribution with Curve')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# 生成二项分布数据
n = 100  # 试验次数
p = 0.3  # 成功概率
x = np.arange(0, n+1)  # 生成 x 轴数据
y = binom.pmf(x, n, p)  # 计算二项分布的概率质量函数值

# 绘制二项分布概率质量函数
plt.bar(x, y, align='center', alpha=0.7)
plt.plot(x, y, color='orange', linestyle='-', linewidth=2, markersize=8, alpha=0.7)
plt.xlabel('Number of Successes')
plt.ylabel('Probability Mass')
plt.title('Binomial Distribution PMF')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import nbinom

# 生成负二项分布数据
r = 10   # 成功次数
p = 0.3  # 成功概率
x = np.arange(r, r+21)  # 生成 x 轴数据
y = nbinom.pmf(x, r, p)  # 计算负二项分布的概率质量函数值

# 绘制负二项分布概率质量函数曲线
plt.plot(x, y, marker='o', color='blue', linestyle='-', linewidth=2, markersize=8)
plt.xlabel('Number of Failures')
plt.ylabel('Probability')
plt.title('Negative Binomial Distribution PMF Curve')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import nbinom

# 参数设置
r = 5   # 成功次数
p = 0.3  # 成功概率
psi = 0.2  # 零膨胀参数
x = np.arange(0, 21)  # 生成 x 轴数据

# 计算零膨胀负二项分布概率质量函数值
y = (1 - psi) * nbinom.pmf(x, r, p) + psi * (x == 0)

# 绘制零膨胀负二项分布概率质量函数曲线
plt.plot(x, y, marker='o', color='blue', linestyle='-', linewidth=2, markersize=8)
plt.xlabel('Number of Failures')
plt.ylabel('Probability')
plt.title('Zero-Inflated Negative Binomial Distribution PMF Curve')
plt.grid()
plt.show()
```
