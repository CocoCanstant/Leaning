# 3. corelation of variables

```python
# the corelation between pclass and survived
import matplotlib.pyplot as plt
import numpy as np

classes = ['Died', 'Survived']
Pclass = ['1st', '2nd', '3rd']

Pclass_survived = {Pclass[0]: np.array([train_data[train_data['Survived'] == 0].Pclass.value_counts()[1] / train_data.Survived.value_counts()[0],
                                        train_data[train_data['Survived'] == 1].Pclass.value_counts()[1] / train_data.Survived.value_counts()[1]]),
                    Pclass[1]: np.array([train_data[train_data['Survived'] == 0].Pclass.value_counts()[2] / train_data.Survived.value_counts()[0],
                                        train_data[train_data['Survived'] == 1].Pclass.value_counts()[2] / train_data.Survived.value_counts()[1]]),
                    Pclass[2]: np.array([train_data[train_data['Survived'] == 0].Pclass.value_counts()[3] / train_data.Survived.value_counts()[0],
                                        train_data[train_data['Survived'] == 1].Pclass.value_counts()[3] / train_data.Survived.value_counts()[1]])}

width = 0.6

fig, ax = plt.subplots()
bottom = np.zeros(2)

index = 0
color_list = ['deepskyblue', 'dodgerblue', 'lightskyblue']

for situation, ration in Pclass_survived.items():
    p = ax.bar(classes, ration, width, label=situation, bottom=bottom, color=color_list[index])
    bottom += ration
    index += 1
    ax.bar_label(p, label_type='center')


ax.set_ylim(0, 1.15)


ax.set_title('Pclass survived situation')
ax.legend(loc='upper left', ncols=3)

plt.show()
```

```python
# the corelation between pclass and survived
classes = ['Died', 'Survived']
Pclass = ['1st', '2nd', '3rd']

Pclass_survived = {classes[0]: np.array([train_data[train_data['Survived'] == 0].Pclass.value_counts()[1] / train_data.Pclass.value_counts()[1],
                                        train_data[train_data['Survived'] == 0].Pclass.value_counts()[2] / train_data.Pclass.value_counts()[2],
                                        train_data[train_data['Survived'] == 0].Pclass.value_counts()[3] / train_data.Pclass.value_counts()[3]]),
                    classes[1]: np.array([train_data[train_data['Survived'] == 1].Pclass.value_counts()[1] / train_data.Pclass.value_counts()[1],
                                        train_data[train_data['Survived'] == 1].Pclass.value_counts()[2] / train_data.Pclass.value_counts()[2],
                                        train_data[train_data['Survived'] == 1].Pclass.value_counts()[3] / train_data.Pclass.value_counts()[3]])}

width = 0.6

fig, ax = plt.subplots()
bottom = np.zeros(3)

index = 0
color_list = ['lightcoral', 'lightskyblue']

for situation, ration in Pclass_survived.items():
    p = ax.bar(Pclass, ration, width, label=situation, bottom=bottom, color=color_list[index])
    bottom += ration
    index += 1
    ax.bar_label(p, label_type='center')


ax.set_ylim(0, 1.15)


ax.set_title('Pclass survived situation')
ax.legend(loc='upper left', ncols=2)

plt.show()
```

```python
# the corelation between sex and survived
sex_list = ['Male', 'Female']

sex_dict = {classes[0]: np.array([train_data[train_data['Survived'] == 0].Sex.value_counts()['male'] / train_data.Sex.value_counts()['male'],
                                  train_data[train_data['Survived'] == 0].Sex.value_counts()['female'] / train_data.Sex.value_counts()['female']]),
            classes[1]: np.array([train_data[train_data['Survived'] == 1].Sex.value_counts()['male'] / train_data.Sex.value_counts()['male'],
                                  train_data[train_data['Survived'] == 1].Sex.value_counts()['female'] / train_data.Sex.value_counts()['female']])}

width = 0.6

fig, ax = plt.subplots()
bottom = np.zeros(2)

index = 0
color_list = ['lightcoral', 'lightskyblue']

for situation, ration in sex_dict.items():
    p = ax.bar(sex_list, ration, width, label=situation, bottom=bottom, color=color_list[index])
    bottom += ration
    index += 1
    ax.bar_label(p, label_type='center')


ax.set_ylim(0, 1.15)


ax.set_title('Pclass survived situation')
ax.legend(loc='upper left', ncols=2)

plt.show()
```
