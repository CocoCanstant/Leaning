```python
pbmc_df = pbmc_df[pbmc_df.文章信息状态 == "待检查"]
print(pbmc_df.shape)
pbmc_df.head(5)
```

```python
pbmc_df["PMID"] = pbmc_df.PMID.astype(int).astype(str)
nogse = pbmc_df[pbmc_df["数据信息"].isna()][["PMID", "Data_url"]]
nogse.reset_index(drop=True, inplace=True)
print(nogse.shape)
nogse.head()
nogse.to_csv('/mnt/d/dailywork/PBMC_database/PMIDwithURL.csv', index=None, encoding='utf-8')
```

```python
gseWithPMID = pbmc_df[~pbmc_df["数据信息"].isna()][["PMID", "数据信息"]]
gseWithPMID.columns = ["PMID", "INFO"]
gseWithPMID.head(5)
```

```python
splitGSE = gseWithPMID.assign(Accession=gseWithPMID.INFO.str.split(",")).explode("Accession")[["PMID", "Accession"]]
```

```python
splitGSE.shape
```

```python
splitGSE["PMID"] = splitGSE.PMID.astype(int)
```

```python
splitGSE["Accession"] = splitGSE.Accession.str.split(';').str[0]
```

```python
splitGSE.reset_index(drop=True, inplace=True)
```

```python
splitGSE["PMID"] = splitGSE["PMID"].astype(str).str.strip()
```

```python
splitGSE["Accession"] = splitGSE["Accession"].str.strip()
```

```python
splitGSE.drop_duplicates(subset=["PMID", "Accession"]).to_csv('/mnt/d/dailywork/PBMC_database/PMIDwithGSE.csv', index=None)
```
