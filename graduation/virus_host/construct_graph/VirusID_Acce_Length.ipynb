```python
import pandas as pd
# Accession and length data
acce_length_data = pd.read_table('/home/ouconstand/Learning/graduation/virus_host/construct_graph/Virion_accession_length.txt', sep='\t', header=None)
acce_length_data.columns = ['Accession', 'Length']
# Accession and VirusID data
acce_virusID_data = pd.read_csv('/home/ouconstand/Learning/graduation/virus_host/construct_graph/final_Acce_VirusID.csv')
```

```python
# get the different accession between two files
ID_accession = set(list(acce_virusID_data['Accession']))
Length_accession = set(list(acce_length_data['Accession']))
print(len(ID_accession))
print(len(Length_accession))
```

```python
# newdata - olddata
New_Old_difference = ID_accession - Length_accession
print("the new unique data includes {}".format(len(New_Old_difference)))
# olddata - newdata
Old_New_difference = Length_accession - ID_accession
print("the old unique data includes {}".format(len(Old_New_difference)))
# intersection area
print("the intersection is {}".format(len(Length_accession & ID_accession)))
# total
print("total number is {}".format(len(Length_accession | ID_accession)))
```
