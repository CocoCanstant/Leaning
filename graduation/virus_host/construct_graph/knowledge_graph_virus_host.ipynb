data process: removing NCBI record is `FALSE` in `host` and `virus` column

```python
import pandas as pd

with open('/home/bio_kang/virus_host_virion/Virion.csv','rb') as f:
    lines = f.readlines()

data_procession = []
for line in lines:
    if len(line) < 34:
        continue
    else:
        data = line.decode('ISO-8859-1').strip('\n').split('\t')

        if data[3] == '' or data[4] == "FALSE" or data[4] == '':
            continue
        else:
            data_procession.append(data)



data = pd.DataFrame(data_procession[1:-1],columns=data_procession[0])
```

extracing coronaviridae data 

```python
import pandas as pd

with open('/home/bio_kang/virus_host_virion/Virion.csv','rb') as f:
    lines = f.readlines()

coronaviridae_data = []

alphacoronavirus = []
betacoronavirus = []
deltacoronavirus =[]
gammacoronavirus =[]

header = lines[0].decode('ISO-8859-1').strip('\n').split('\t')

for line in lines:
    line = line.decode('ISO-8859-1').strip('\n').split('\t')
    if len(line) < 34:
        continue
    else:
        if line[13] == 'coronaviridae':
            coronaviridae_data.append(line)

            if line[12] == 'alphacoronavirus':
                if line[33] != '':
                    alphacoronavirus.append(line)
            elif line[12] == 'betacoronavirus':
                if line[33] != '':
                    betacoronavirus.append(line)
            elif line[12] == 'deltacoronavirus':
                if line[33] != '':
                    deltacoronavirus.append(line)
            elif line[12] == 'gammacoronavirus':
                if line[33] != '':
                    gammacoronavirus.append(line)

data = pd.DataFrame(coronaviridae_data, columns=header)

alpha = pd.DataFrame(alphacoronavirus, columns=header)
beta = pd.DataFrame(betacoronavirus, columns=header)
delta = pd.DataFrame(deltacoronavirus, columns=header)
gamma = pd.DataFrame(gammacoronavirus, columns=header)

alpha.to_csv('alphacoronavirus.csv',index=False, sep='\t')
beta.to_csv('betacoronavirus.csv',index=False, sep='\t')
delta.to_csv('deltacoronavirus.csv',index=False, sep='\t')
gamma.to_csv('gammacoronavirus.csv',index=False, sep='\t')

data.to_csv('coronaviridae_virus.csv',index=False)
```

```python
import re
import requests
import os

def from_accession_to_id(segment):
    term_url = 'https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esearch.fcgi?db=nucleotide&api_key=6e2f58038b074d964bb8b87f43623af7b508&term=' + ",".join(segment)
    term_info = requests.get(term_url, timeout=(5,5)).text
 
    id_pattern = re.compile(r'<Id>(\d+)</Id>')
    accession_id = re.findall(id_pattern, term_info)
    return accession_id

def from_id_to_version_accession(id):
    fasta_url = 'https://eutils.ncbi.nlm.nih.gov/entrez/eutils/efetch.fcgi?db=nucleotide&rettype=fasta&id={}&api_key=6e2f58038b074d964bb8b87f43623af7b508'.format(','.join(id))
    fasta_info = requests.get(fasta_url,timeout=(5,5)).text

    version_pattern = re.compile(r'>[A-Z]{1,2}\d+.\d')
    version_info = re.findall(version_pattern, fasta_info)
    return version_info


genus_set = ["alphacoronavirus","betacoronavirus","gammacoronavirus","deltacoronavirus"]

for genus in genus_set:
    with open('non_version_{}.txt'.format(genus), 'r') as f:
        f_lines = f.readlines()
    with open('pure_num_{}.txt'.format(genus),'r') as k:
        k_lines = k.readlines()

    f_data = list(map(lambda x:x.strip('\n'), f_lines))
    f_time = len(f_data) // 20 + 1

    k_data = list(map(lambda x:x.strip('\n'),k_lines))
    k_time = len(k_data) // 20 +1

    version = []
    
    for i in range(f_time):
        segment = f_data[i*20:(i+1)*20]
        id = from_accession_to_id(segment)
        version = version + from_id_to_version_accession(id)

    for j in range(k_time):
        id = k_data[i*20:(i+1)*20]
        version = version + from_id_to_version_accession(id)

    with open('including_version_{}.txt'.format(genus), 'w') as m:
        m.writelines(list(map(lambda x:x+'\n',version)))

    os.system("grep -oP '[A-Z]+\d+.\d' including_version_{}.txt >> criterion_format_{}.txt".format(genus,genus))
    os.system("sort -u criterion_format_{}.txt > finally_{}.txt".format(genus,genus))
```
