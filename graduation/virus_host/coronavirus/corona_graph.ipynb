```python
import os

os.system(' cat host_id.txt | /home/ouyangkang/software/anaconda3/envs/bio/bin/taxonkit reformat -I 1 -f "{p};{c};{o};{f};{g};{s}" > host_taxon.txt ')
```

```python
# translate the format
with open('host_taxon.txt', 'r') as f:
    data = f.readlines()

host_tax = {}

for line in data:
    line = line.strip('\n').split('\t')
    taxon = line[1].split(';')
    host_tax[line[0]] = taxon
```

```python
import pandas as pd

host_data = pd.DataFrame(host_tax).T
host_data = host_data.reset_index()
host_data.columns = ['Host_id','Phylum','Class','Order','Family','Genus','Species']
display(host_data)
host_data.to_csv('host_taxon.csv', index=None)
```
